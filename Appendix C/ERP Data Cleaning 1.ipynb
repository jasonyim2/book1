# C.4.4 기타 변수 데이터처리 1

```python
df["CapitalRatio1"].isnull().sum() # CapitalRatio1값 중에 null값의 갯수 체크. 결과는 0개로 나옴
```

```python
df["CapitalRatio1"].value_counts(dropna=False) # CapitalRatio1값을 종류별로 다 세어서 결과를 체크
```

```python
df1 = df.copy()   # 원본 데이터프레임 df를 복사(copy)하여 다른 데이터프레임 이름인 df1으로 저장 
                  # 이러면 원래 데이터프레임 df의 변형을 막는 효과가 있음.

# 새로 명명한 데이터프레임 df1의 CapitalRatio1의 변수값 중에서 %를 포함하지 않는 그룹을 df_part1에 저장    
df_part1 = df1[df1["CapitalRatio1"].str.contains('%') == False]
df_part1.shape # row갯수를 확인하여 결과는 79,342개로 나옴.
```

```python
# 새로 명명한 데이터프레임 df1의 CapitalRatio1의 변수값 중에서 %를 포함하는 그룹을 df_part2에 저장    
df_part2 = df1[df1["CapitalRatio1"].str.contains('%') == True]
df_part2.shape # row갯수를 확인하여 결과는 20,905개로 나옴
               # 이로서 dfpart1과 dfpart2를 합치면 원래 df 갯수인 100,247개가 도출
```

```python
df_part1['CapitalRatio1'].value_counts(dropna=False) # 점(.) 갯수가 3,690개로 도출
```

```python

mask = pd.to_numeric(df_part1['CapitalRatio1'], errors='coerce').isna()
a = mask.sum() 
print (a)  # 점(.) 갯수가 3,690개임을 재확인
```

```python
# 점(.)을 np.nan으로 변환
df_part1["CapitalRatio1"] = df_part1["CapitalRatio1"].replace('.', np.nan)

# null value를 0으로 변환하고 그후 dtype을 int64로 변환
df_part1["CapitalRatio1"] = df_part1["CapitalRatio1"].fillna(0).astype(np.int64) 
df_part1['CapitalRatio1'].value_counts(dropna=False) # dtype이 int64로 바뀌었음을 확인.
```

```python
df_part1['CapitalRatio1'].describe() # 최소값과 최대값을 확인하여 CapitalRatio1값의 범위를 확인
```

```python
df_part2['CapitalRatio1'].value_counts(dropna=False)
```

```python
# CapitalRatio1 변수값의 %를 빈칸(blank)으로 바꾸고, dtype을 실수(float)로 바꾸고 100을 곱함 

df_part2['CapitalRatio1'] = df_part2['CapitalRatio1'].str.replace(r'%', r'').astype('float')*100
df_part2['CapitalRatio1'].value_counts(dropna=False)
```

```python
df_part2['CapitalRatio1'].describe() # 최소값과 최대값을 확인하여 CapitalRatio1값의 범위를 확인
```

```python
# 두개로 분리된 원 데이터프레임 파일들을 다시 하나로 병합.

dfu = df_part1.append(df_part2, ignore_index = True)
dfu.shape #분할전 row 갯수인 100,247개 다시 회복
```

```python
dft = dfu.drop('CapitalRatio1', axis=1)   # 데이터프레임 dfu에서 CapitalRatio1 변수만 제거
list(dft.columns)   # 컬럼명 확인
```

```python
dft = dft.replace('.', np.nan)   # dft내에 있는 모든 변수들의 값 중에 점(.)을 np.nan으로 대체  
```

```python
dft['CapitalRatio1'] = dfu['CapitalRatio1']
list(dft.columns)
```
