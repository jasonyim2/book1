# C.5.6 t 검정(옵션)

```python
from scipy import stats

data_1 = df[df['EBizSystem2'] == 1]['RevProfit2']
data_0 = df[df['EBizSystem2'] == 0]['RevProfit2']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue < 0.05 
```

```python
from scipy import stats

data_1 = df[df['EBizSystem2'] == 1]['Liability2']
data_0 = df[df['EBizSystem2'] == 0]['Liability2']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue < 0.05 
```

```python
from scipy import stats

data_1 = df[df['EBizSystem2'] == 1]['Emp_total']
data_0 = df[df['EBizSystem2'] == 0]['Emp_total']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue < 0.05 
```
