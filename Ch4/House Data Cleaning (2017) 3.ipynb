# 4.5.6 t 검정

```python
from scipy import stats

data_1 = df[df['VALP_B1'] == 1]['HINCP']
data_0 = df[df['VALP_B1'] == 0]['HINCP']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue < 0.05 
```

```python
data_1 = df[df['VALP_B1'] == 1]['AGEP']
data_0 = df[df['VALP_B1'] == 0]['AGEP']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue > 0.05    
```

```python
data_1 = df[df['VALP_B1'] == 1]['BDSP']
data_0 = df[df['VALP_B1'] == 0]['BDSP']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue < 0.05    
```

```python
data_1 = df[df['VALP_B1'] == 1]['ELEP']
data_0 = df[df['VALP_B1'] == 0]['ELEP']

stats.ttest_ind(data_1, data_0)    # 결과는 pvalue < 0.05     
```

```python
data_1 = df[df['VALP_B1'] == 1]['GASP']
data_0 = df[df['VALP_B1'] == 0]['GASP']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue > 0.05    
```

```python
data_1 = df[df['VALP_B1'] == 1]['NRC']
data_0 = df[df['VALP_B1'] == 0]['NRC']

stats.ttest_ind(data_1, data_0)    # p-value가 과학적 표기임에 주의, # 결과는 pvalue < 0.05  
```

```python
data_1 = df[df['VALP_B1'] == 1]['RMSP']
data_0 = df[df['VALP_B1'] == 0]['RMSP']

stats.ttest_ind(data_1, data_0)   # 결과는 pvalue < 0.05    
```
